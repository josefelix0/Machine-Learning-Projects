```python
# idetify missing data
df.isnull().sum()
```

```python
# Fill Missing data values ::
Cate_features = ["age" , "embarked" , "deck" , "embark_town"]
for col in Cate_features :
    df[col].fillna(df[col].mode()[0] , inplace=True)
```

```python
df.isnull().sum()
```

```python
# detect outliers and replace ::
from datasist.structdata import detect_outliers
cols = df.select_dtypes(include=["number"]).columns
for col in cols :
    ouliers = detect_outliers(df , 0 , [col])
    med_value = df[col].median()
    df[col].iloc[ouliers] = med_value
```

```python
cols = df.select_dtypes(include=["number"]).columns
for col in cols :
    df[col] = df[col].apply(lambda x : round(x))
```

```python
df
```

```python
# Handlling Text data
df = pd.get_dummies(df , drop_first=True)
```

```python
df
```

```python
# Split data into input data variables and Target
x = df.drop( "alive_yes" , axis=1)
y = df["alive_yes"]
```

* import Requried liberaries and Framworks

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split , cross_val_score
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
# split data into train and test data
x_train , x_test , y_train , y_test = train_test_split(x , y , random_state=20 , test_size=0.2 , shuffle=True)
```

```python
# Normalize data
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test  = scaler.transform(x_test)
```

```python
# Bulid models 
models = {"RF_Model" : RandomForestClassifier(n_estimators = 100 , random_state = 20) ,
          "XGC" : XGBClassifier() }
```

```python
# comparing among 2 Models 
for name , model in models.items() :
    print(f"The Model :: {name}\n")
    model.fit(x_train , y_train)
    train_score = model.score(x_train , y_train)
    test_score = model.score(x_test , y_test)
    y_pred = model.predict(x_test)
    evalute = ds.model.get_classification_report(y_pred , y_test)

    print(f"The Train_score :: {train_score}\nThe Test_score :: {test_score}\n")
    print(f"The y_pre :: {y_pred}\n-------------")
    print(f"The Evaluate model ::\n{evalute}\n--------------------\n")
    print("_" * 80)
    
    
```
